```python
import keras

import matplotlib.pyplot as plt
import cv2
import numpy as np
import os

import random
import pickle

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,MaxPooling2D,Conv2D

model_main=keras.models.load_model("model_multiclassification_final")
def predict_Video(frame):
    # try_data_path = path
    # try_data = cv2.imread(frame, cv2.IMREAD_GRAYSCALE)
    # Resizing the image and reshaping the image so it can pass the Convolutional Layer as a 4D arrays
    resized_try_data = cv2.resize(frame, (250, 250))
    try_data_resized = resized_try_data.reshape(1, 250, 250, 3)# Check the shape of the resized image arrays
    # print(try_data_resized.shape)

#     plt.imshow(try_data_resized[0], cmap='gray')
    # Declare the new data as Xnew
    Xnew = try_data_resized
    # make a prediction using the trained models
    ynew = model_main.predict(Xnew)
    print(ynew)
    return ynew
    # print(type(ynew[0])
    # show the predicted Outputs


```

```python
import matplotlib as mpl
import matplotlib.cm as mtpltcm
cap=cv2.VideoCapture(0)
# cap_2=cv2.VideoCapture(0)
# cv2.setWindowProperty(cap_2,cv2.WINDOW_NORMAL)

font1 = cv2.FONT_HERSHEY_SIMPLEX
# org for the text being specified
org1 = (50, 50)
# font scale for the text being specified
fontScale1 = 0.8
# Blue color for the text being specified from BGR
color1 = ( 0 , 255 , 0 )
# Line thickness for the text being specified at 2 px
thickness1 = 2
# Using the cv2.putText() method for inserting text in the image of the specified path



while(cap.isOpened()):
    ret,frame=cap.read()
    frame=cv2.resize(frame,(250,250))
    colormap = mpl.cm.jet
    cNorm = mpl.colors.Normalize(vmin=0, vmax=255)
    scalarMap = mtpltcm.ScalarMappable(norm=cNorm, cmap=colormap)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #add blur to make it more realistic
#         blur = cv2.GaussianBlur(gray,(15,15),0)
        #assign colormap
    colors = scalarMap.to_rgba(gray, bytes=False)

    cv2.imshow("Image",frame)
    cv2.imshow("Gray",colors)
    returned_result=predict_Video(frame)
    # print(returned_result)
    if np.argmax(returned_result,axis=1) == 1:
        cv2.putText(frame, 'Human Detected', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)
#         cv2.imshow('Image_2',colors)
        
    elif np.argmax(returned_result,axis=1) == 0:
        cv2.putText(frame, 'Chair', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)
    
    elif np.argmax(returned_result,axis=1) == 3:
        cv2.putText(frame, 'Sofa', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)
    
    
    elif np.argmax(returned_result,axis=1) == 2:
        cv2.putText(frame, 'Bed', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)
    
    
    elif np.argmax(returned_result,axis=1) == 4:
        cv2.putText(frame, 'Tabel', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)


    else:
        cv2.putText(frame, 'None Detected', org1, font1,
        fontScale1, color1, thickness1, cv2.LINE_AA)
        cv2.imshow("Image",frame)
        # print("Predicted=%s" % ('Not Detected'))

    if cv2.waitKey(1)==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
